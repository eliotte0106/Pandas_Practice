# 4. File Save/Open
DataFrame with excel, csv, or other files

```python
import pandas as pd

data = {
    'name' : ['Park', 'Lee', 'Kim', 'Rishi', 'Chen', 'Meggie'],
    'school' : ['GATech','GATech','GATech','GATech','GATech', 'GATech'],
    'language' : ['Python', 'Java', 'C', 'Ruby', 'TypeScript', ''],
    'gpa' : [3.4, 3.2, 3.8, 4.0, 3.5, 3.1]
}
df = pd.DataFrame(data, index = ['1', '2', '3', '4', '5', '6'])
df.index.name = 'number'
df
```

## Save
### save as different types of files

```python
df.to_csv('score.csv') # if doesn't work on excel, add 'score.csv', encoding='utf-8-sig'
```

```python
df.to_csv('score.txt', sep = '\t') # sep by tab, txt file
```

```python
df.to_excel('score.xlsx')
```

***
## Open

```python
df = pd.read_csv('score.csv')
df
```

```python
df = pd.read_csv('score.csv', skiprows = 1) # skip the number of rows
df
```

```python
df = pd.read_csv('score.csv', skiprows = [1, 3, 5]) # 1, 3, 5 skip
df
```

```python
df = pd.read_csv('score.csv', nrows = 4) # first 4 rows
df
```

```python
df = pd.read_csv('score.csv', skiprows = 2, nrows = 3)
df
```

```python
df = pd.read_csv('score.txt', sep = '\t')
df
```

```python
df = pd.read_excel('score.xlsx')
df
```
