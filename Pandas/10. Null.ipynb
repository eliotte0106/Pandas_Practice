# 10. Null

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col = 'number')
df
```

## fill na

```python
df.fillna('no') # replace NaN with no
```

```python
import numpy as np
df['language'] = np.nan # change language to NaN
df
```

```python
df.fillna('undecided', inplace = True)
df
```

```python
df = pd.read_excel('score.xlsx', index_col = 'number')
df
```

```python
df['language'].fillna('JavaScript', inplace = True)
df
```

## dropna

```python
df = pd.read_excel('score.xlsx', index_col = 'number')
df
```

```python
df.dropna() # delete data contains NaN
df
```

```python
df = pd.read_excel('score.xlsx', index_col = 'number')
```

- axis: index or column
- how: any or all

```python
df.dropna(axis = 'index', how = 'any') # delete row if row has 1 or more NaN
```

```python
df
```

```python
df.dropna(axis = 'columns') # delete column if column has 1 or more NaN
```

```python
df.dropna(axis = 'columns', how = 'all') # delete if all data is NaN
```
